```python
% matplotlib inline

import numpy as np
import numpy.ma as ma
import pandas as pd
from math import *
import matplotlib.pyplot as plt
import scipy.io as sio
from mpl_toolkits.basemap import Basemap, cm

import MITgcmutils as mit
```

```python
#import grid params for CR grid

in_dir = '/Users/archie/Google Drive/work/comparison/data/'

in_file = '%sCRgrid/HFacC.data' % in_dir
HFCR = np.fromfile(in_file, dtype = '>f')
HFCR = HFCR.reshape(23, 160, 360)[0,:,:]

in_file = '%sCRgrid/XC.data' % in_dir
XC = np.fromfile(in_file, dtype = '>f')
XC = XC.reshape(160,360)

in_file = '%sCRgrid/YC.data' % in_dir
YC = np.fromfile(in_file, dtype = '>f')
YC = YC.reshape(160,360)
```

```python
# import  data

in_file = '%sannual/HR2CR/HR2CRmld.1999.data' % in_dir
HRmld = np.fromfile(in_file, dtype='>f').reshape(160, 360)
HRmask = ma.masked_where(HFCR == 0, HRmld, copy=False)

```

```python
palette = plt.get_cmap('RdYlBu_r', 50)
plt.pcolormesh(XC, YC, HRmask,shading='flat', vmin= 0, vmax = 250, cmap = palette)
plt.axis([0, 360, -80, 80])
plt.colorbar()
plt.show()

plt.plot(np.nanmean(HRmld,1), YC[:,0])
plt.axis([0, 100, -60, 60])
plt.show()
```
